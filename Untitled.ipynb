```python
df_infill_pattern = df[['infill_pattern']]
df_material = df[['material']]
```

```python
one_hot_encoder = OneHotEncoder()
df_pattern_encoded = one_hot_encoder.fit_transform(df_infill_pattern)
df_pattern_encoded.toarray()
```

```python
pattern_encoded=pd.DataFrame(df_pattern_encoded.toarray(),columns=df['infill_pattern'].unique())
```

```python
one_hot_encoder = OneHotEncoder()
df_material_encoded = one_hot_encoder.fit_transform(df_material)
df_material_encoded.toarray()
```

```python
material_encoded=pd.DataFrame(df_material_encoded.toarray(),columns=df['material'].unique())
```

```python
material_encoded
```

```python
df_final=pd.merge(df,pattern_encoded,right_index=True,left_index=True)
```

```python
df_final=pd.merge(df_final,material_encoded,right_index=True,left_index=True)
```

```python
df_final.head()
```
